```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt 
import netCDF4 as nc
from netCDF4 import Dataset
import pyproj
import pandas as pd

#ncdump 'PhysStates.nc' -h in terminal 
```

```python
#importing SINMOD data 
filename_PhysStates = 'PhysStates.nc'
filename_BioStates = 'BioStates.nc'

PhysStates_data = Dataset(filename_PhysStates, 'r')
BioStates_data = Dataset(filename_BioStates, 'r')

xc = PhysStates_data.variables['xc'][:]  # x-coordinates (meters)
yc = PhysStates_data.variables['yc'][:]  # y-coordinates (meters)


#Phys states:
elevation = PhysStates_data['elevation'][0,:,:] #first time slice
temperature = PhysStates_data['temperature'][0,:,:,:]  
salinity = PhysStates_data['salinity'][0,:,:,:] 
v_velocity = PhysStates_data['v_velocity'][0,:,:,:]  
u_velocity = PhysStates_data['u_velocity'][0,:,:,:]  
depth_phys = PhysStates_data['depth'][:]
time_phys = PhysStates_data['time'][:]

#Bio states:
time_bio = BioStates_data['time'][:]
depth_bio = BioStates_data['depth'][:]
nitrate = BioStates_data['nitrate'][0,:,:,:] #first time slice 
silicate = BioStates_data['silicate'][0,:,:,:]
ammonium = BioStates_data['ammonium'][0,:,:,:]
diatoms = BioStates_data['diatoms'][0,:,:,:]
flagellates = BioStates_data['flagellates'][0,:,:,:]
ciliates = BioStates_data['ciliates'][0,:,:,:]
hnano = BioStates_data['HNANO'][0,:,:,:]
bacteria= BioStates_data['bacteria'][0,:,:,:]
calanus_finmarchicus = BioStates_data['calanus_finmarchicus'][0,:,:]
calanus_glacialis = BioStates_data['calanus_glacialis'][0,:,:]
detritus_slow = BioStates_data['detritus_slow'][0,:,:,:]
detritus_fast = BioStates_data['detritus_fast'][0,:,:,:]
doc = BioStates_data['DOC'][0,:,:,:]         #dissolved organic carbon
cDOM =BioStates_data['cDOM'][0,:,:,:]        #colored dissolved organic matter 
silicate_detritus = BioStates_data['silicate_detritus'][0,:,:,:]
sediment_Si = BioStates_data['sediment_Si'][0,:,:]   #Sediment silica (Si) 
sediment_N = BioStates_data['sediment_N'][0,:,:]     #Sediment nitrogen


#Grid mapping (same for both)
grid_mapping = PhysStates_data['grid_mapping']
```

```python
z_levels, y_dim, x_dim = temperature.shape

# Reshape temperature array into 2D: (z, y*x) where y*x are the coordinates
temperature_reshaped = temperature.reshape(z_levels, y_dim * x_dim)

# Create column labels representing (x, y) coordinates
column_labels = [(x, y) for y in range(y_dim) for x in range(x_dim)]

# Create the DataFrame, with z indexed along rows, and (x, y) coordinates as columns
temperature_df = pd.DataFrame(temperature_reshaped, columns=pd.MultiIndex.from_tuples(column_labels, names=["x", "y"]))

# Optionally set the index for the z levels
temperature_df.index.name = "z"

# Display the DataFrame
print(temperature_df)
```

```python
#importing coral reef data

gml_file_coral_reefs = "/Users/maikentomren/Documents/prosjektoppgave/plotting_first_tries/KystOgFiskeri_50_Trondelag_25832_Korallrev_GML.gml"
gdf_coral_reefs = gpd.read_file(gml_file_coral_reefs)
coral_data = gdf_coral_reefs[gdf_coral_reefs['naturtypeNavn'] == 'Korallforekomster']
coral_df = pd.DataFrame(coral_data)
coral_location_df = coral_df[['lengdegrad', 'breddegrad']]

```

```python
#specifying projection to SINMOD format

#grid mapping info in SINMOD file
#print(grid_mapping)

from pyproj import CRS, Transformer

# Define the Coordinate Reference Systems (CRS)
crs_wgs84 = CRS.from_epsg(4326)  # WGS84 (lat/lon coordinates)
crs_euref89_utm32 = CRS.from_epsg(25832)  # EUREF89 / UTM zone 32N
crs_euref89 = CRS.from_epsg(4258)  # EUREF89 (geographic lat/lon)

# SINMOD projection parameters as a custom projection
crs_sinmod = CRS.from_proj4("+proj=stere +lat_0=90 +lat_ts=60 +lon_0=58 "
                            "+x_0=3304000 +y_0=2554000 +ellps=WGS84 +units=m +no_defs")


# Create a transformer to transform from WGS84 to SINMOD
transformer_wgs84_to_sinmod = Transformer.from_crs(crs_wgs84, crs_sinmod, always_xy=True)

# Create a transformer to transform from EUREF89 (geographic) to SINMOD
transformer_euref89_to_sinmod = Transformer.from_crs(crs_euref89, crs_sinmod, always_xy=True)

transformer_sinmod_to_wgs84 = Transformer.from_crs(crs_sinmod, crs_wgs84, always_xy=True)

xc_grid, yc_grid = np.meshgrid(xc, yc)
lon_grid, lat_grid = transformer_sinmod_to_wgs84.transform(xc_grid, yc_grid)

# Performing projection on coral reef data

print("coral_lons range before:", coral_location_df['lengdegrad'].min(), coral_location_df['lengdegrad'].max())
print("coral_lats range before:", coral_location_df['breddegrad'].min(), coral_location_df['breddegrad'].max())
print("--------------------------------")
# Transform coral coordinates from EUREF89 to SINMOD
coral_x, coral_y = transformer_euref89_to_sinmod.transform(coral_location_df['lengdegrad'], coral_location_df['breddegrad'])
coral_x /= 800
coral_y /= 800
print("coral_lons range after:", coral_x.min(), coral_x.max())
print("coral_lats range after:", coral_y.min(), coral_y.max())
print("--------------------------------")
print("Grid x values range:", xc_grid.min(), xc_grid.max())
print("Grid y values range:",yc_grid.min(), yc_grid.max())
print("--------------------------------")
print("Grid lon values range:", lon_grid.min(), lon_grid.max())
print("Grid lat values range:", lat_grid.min(), lat_grid.max())


```
